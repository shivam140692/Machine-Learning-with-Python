```python
# fill the null value of Age column with the mean values of each Pclass category
# for multiple columns we use a square bracket additional to the dataframe ka square bracket
#inpute_age is the function name for two or  more than 2 params of list
# when you are defining two columns or passing two values in the lamba function then use axis=1


def inpute_age(list):
    age= list[0]     # age and pclass here is a variable we are defining
    pclass= list[1]
    
    if pd.isnull(age):
        
        if pclass ==1:
            return 38
        elif pclass ==2:
            return 30
        else:
            return 25
    else:
        return age
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(inpute_age, axis=1)
```

```python
# adding columns makes datframe less complex
# get_dummies function will convert the column values into numbers form of 1 or 0 that will give you the dataframe
# for sex, male =1 and female = 0
# Embarked has three values so it will give the combination as
# s C Q
# 1 0 0
# 0 1 0
# 0 0 1
# so if we eliminate the one column among the (S, c and Q) still will get the unique values

Sex= pd.get_dummies(df['Sex'], drop_first=True)
Embarked= pd.get_dummies(df['Embarked'], drop_first=True)
```

```python
df.drop(['Sex', 'Embarked'], axis=1, inplace=True)
```

```python
df = pd.concat([df,Sex, Embarked],axis=1)
```
